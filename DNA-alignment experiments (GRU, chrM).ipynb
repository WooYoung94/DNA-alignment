```python
fig, ax = plt.subplots(1, 1, figsize = (17, 9))

ax.plot(gru1, label = 'test1') # gru : 4 -> 128, fc : 128 -> 1
ax.plot(gru2, label = 'test2') # gru : 4 -> 128, fc : 128 -> 64 -> 1 ( fc  non-linear  )
ax.plot(gru3, label = 'test3') # gru : 4 -> 256, fc : 256 -> 64 -> 1 ( fc  non-linear  )
ax.plot(gru4, label = 'test4') # gru : 4 -> 256, fc : 256 -> 64 -> 1 (fc  relu)
ax.plot(gru5, label = 'test5') # gru : 4 -> 256, fc : 256 -> 64 -> 1, batch : 1024 -> 5120 (  lstm    )
#gru5  epcoh     
ax.plot(gru6, label = 'test6') # gru : 4 -> 256, fc : 256 -> 64 -> 1, batch : 1024, layer : 2
ax.plot(gru7, label = 'test7') # gru : 4 -> 256, fc : 256 -> 64 -> 1, batch : 1024, layer : 2, dropout : 0.1
ax.plot(gru8, label = 'test8') # gru7  sigmoid  clamp ,  
ax.plot(gru9, label = 'test9') # gru7  sigmoid * 0.8 + 0.1  idx  
ax.plot(gru10, label = 'test10') # gru7  sigmoid * 0.6 + 0.2  idx  
ax.plot(gru11, label = 'test11') # gru 7  bidirectional 
plt.axhline(y = 0.0, c = 'k', linewidth = 0.5)
ax.plot()
plt.legend()
plt.savefig('./fig.png')
plt.show()
```
