```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('log/gru/loss7.txt', 'r') as fs:
    
    gru7 = eval(fs.read())
```

```python
with open('log/gru/loss11.txt', 'r') as fs:
    
    gru11 = eval(fs.read())
```

```python
with open('log/cnn/loss1.txt', 'r') as fs:
    
    cnn1 = eval(fs.read())
```

```python
fig, ax = plt.subplots(1, 1, figsize = (17, 9))

ax.plot(gru7, label = 'gru') # gru : 4 -> 256, fc : 256 -> 64 -> 1, batch : 1024, layer : 2, dropout : 0.1
ax.plot(gru11, label = 'gru11') # gru 7 에서 bidirectional 추가
ax.plot(cnn1, label = 'cnn1') # gru 7 에서 bidirectional 추가
plt.axhline(y = 0.0, c = 'k', linewidth = 0.5)
ax.plot()
plt.legend()
plt.savefig('./fig.png')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (17, 9))

ax.plot(gru7[1500:], label = 'gru7') # gru : 4 -> 256, fc : 256 -> 64 -> 1, batch : 1024, layer : 2, dropout : 0.1
ax.plot(gru11[1500:], label = 'gru11') # gru7 에서 bidirectional 추가
ax.plot(cnn1[1500:], label = 'cnn1') # gru7 에서 bidirectional 추가
plt.axhline(y = 0.0, c = 'k', linewidth=0.5)
ax.plot()
plt.legend()
plt.savefig('./fig.png')
plt.show()
```
