```python
fig, ax = plt.subplots(1, 1, figsize = (17, 9))

ax.plot(gru1, label = 'test1') # gru : 4 -> 128, fc : 128 -> 1
ax.plot(gru2, label = 'test2') # gru : 4 -> 128, fc : 128 -> 64 -> 1 (실수로 fc 에 non-linear 없이 함)
ax.plot(gru3, label = 'test3') # gru : 4 -> 256, fc : 256 -> 64 -> 1 (실수로 fc 에 non-linear 없이 함)
ax.plot(gru4, label = 'test4') # gru : 4 -> 256, fc : 256 -> 64 -> 1 (fc 사이에 relu)
ax.plot(gru5, label = 'test5') # gru : 4 -> 256, fc : 256 -> 64 -> 1, batch : 1024 -> 5120 (이 세팅의 lstm 은 학습이 매우 느림)
#gru5 세팅은 epcoh 늘려서 다시 시도할 필요 있음
ax.plot(gru6, label = 'test6') # gru : 4 -> 256, fc : 256 -> 64 -> 1, batch : 1024, layer : 2
ax.plot(gru7, label = 'test7') # gru : 4 -> 256, fc : 256 -> 64 -> 1, batch : 1024, layer : 2, dropout : 0.1
ax.plot(gru8, label = 'test8') # gru7 에서 sigmoid 대신 clamp 함, 학습 실패
ax.plot(gru9, label = 'test9') # gru7 에서 sigmoid * 0.8 + 0.1 을 idx 로 사용
ax.plot(gru10, label = 'test10') # gru7 에서 sigmoid * 0.6 + 0.2 을 idx 로 사용
plt.axhline(y = 0.0, c = 'k', linewidth=0.5)
ax.plot()
plt.legend()
plt.savefig('./fig.png')
plt.show()
```
