```python
import os
import sys
import pickle
import numpy as np
from sklearn.manifold import TSNE
from sklearn.manifold import MDS
from sklearn.manifold import Isomap
from sklearn.manifold import SpectralEmbedding
import matplotlib.pyplot as plt
```

```python
with open('/home/youngwoo/Desktop/yoon_kim_cnn1_latent_origin.pickle', 'rb') as fs:
    
    origin = np.array(pickle.load(fs))
```

```python
with open('/home/youngwoo/Desktop/yoon_kim_cnn1_latent_per.pickle', 'rb') as fs:
    
    per = np.array(pickle.load(fs))
```

```python
idxs = np.arange(len(origin))
```

```python
embedding_tsne_origin = TSNE(n_components = 2).fit_transform(origin)
plt.scatter(embedding_tsne_origin[:,0], embedding_tsne_origin[:,1], c = idxs)
plt.show()
```

```python
embedding_mds_origin = MDS(n_components = 2, n_jobs = 20).fit_transform(origin)
plt.scatter(embedding_mds_origin[:,0], embedding_mds_origin[:,1], c = idxs)
plt.show()
```

```python
embedding_isomap_origin = Isomap(n_neighbors = 5, n_components = 2, n_jobs = 20).fit_transform(origin)
plt.scatter(embedding_isomap_origin[:,0], embedding_isomap_origin[:,1], c = idxs)
plt.show()
```

```python
embedding_spectral_origin = SpectralEmbedding(n_components = 2, n_jobs = 20).fit_transform(origin)
plt.scatter(embedding_spectral_origin[:,0], embedding_spectral_origin[:,1], c = idxs)
plt.show()
```

```python
embedding_tsne_per = TSNE(n_components = 2).fit_transform(per)
plt.scatter(embedding_tsne_per[:,0], embedding_tsne_per[:,1], c = idxs)
plt.show()
```

```python
embedding_mds_per = MDS(n_components = 2, n_jobs = 20).fit_transform(per)
plt.scatter(embedding_mds_per[:,0], embedding_mds_per[:,1], c = idxs)
plt.show()
```

```python
embedding_isomap_per = Isomap(n_neighbors = 5, n_components = 2, n_jobs = 20).fit_transform(per)
plt.scatter(embedding_isomap_per[:,0], embedding_isomap_per[:,1], c = idxs)
plt.show()
```

```python
embedding_spectral_per = SpectralEmbedding(n_components = 2, n_jobs = 20).fit_transform(per)
plt.scatter(embedding_spectral_per[:,0], embedding_spectral_per[:,1], c = idxs)
plt.show()
```
