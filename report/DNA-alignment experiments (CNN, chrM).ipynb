```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('../log/gru/loss7.txt', 'r') as fs:
    
    gru7 = eval(fs.read())
```

```python
with open('../log/gru/loss11.txt', 'r') as fs:
    
    gru11 = eval(fs.read())
```

```python
with open('../log/cnn/loss1.txt', 'r') as fs:
    
    cnn1 = eval(fs.read())
```

```python
with open('../log/cnn/loss2.txt', 'r') as fs:
    
    cnn2 = eval(fs.read())
```

```python
with open('../log/cnn/loss3.txt', 'r') as fs:
    
    cnn3 = eval(fs.read())
```

```python
fig, ax = plt.subplots(1, 1, figsize = (17, 9))

ax.plot(gru7, label = 'gru7') # gru : 4 -> 256, fc : 256 -> 64 -> 1, batch : 1024, layer : 2, dropout : 0.1
ax.plot(gru11, label = 'gru11') # gru 7  bidirectional 
ax.plot(cnn1, label = 'cnn1') # one hot  64ch 1d  ,  64 -> 128 -> 256 -> 512 -> 1024 -> 1  relu  conv 
ax.plot(cnn2, label = 'cnn2') # cnn1  data augmentation  W  
#ax.plot(cnn3, label = 'cnn3') # cnn1  data augmentation , reparam ,  
plt.axhline(y = 0.0, c = 'k', linewidth = 0.5)
ax.plot()
plt.legend()
plt.savefig('./fig.png')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (17, 9))

ax.plot(gru7[1500:], label = 'gru7') # gru : 4 -> 256, fc : 256 -> 64 -> 1, batch : 1024, layer : 2, dropout : 0.1
ax.plot(gru11[1500:], label = 'gru11') # gru7  bidirectional 
ax.plot(cnn1[1500:], label = 'cnn1') # one hot  64ch 1d  ,  64 -> 128 -> 256 -> 512 -> 1024 -> 1  relu  conv 
ax.plot(cnn2[1500:], label = 'cnn2') # cnn1  data augmentation  W  
#ax.plot(cnn3[1500:], label = 'cnn3') # cnn1  data augmentation , reparam ,  
plt.axhline(y = 0.0, c = 'k', linewidth=0.5)
ax.plot()
plt.legend()
plt.savefig('./fig.png')
plt.show()
```
