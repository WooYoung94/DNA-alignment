```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('../log/cnn2/loss1.txt', 'r') as fs:
    
    cnn1 = eval(fs.read())
```

```python
with open('../log/cnn2/loss2.txt', 'r') as fs:
    
    cnn2 = eval(fs.read())
```

```python
fig, ax = plt.subplots(1, 1, figsize = (17, 9))

ax.plot(cnn1, label = 'cnn1') # 기본 cnn 모델에 1% augmented (저번 최종 결과)
ax.plot(cnn2, label = 'cnn2') # 기본 cnn 모델에 15% augmented
plt.axhline(y = 0.0, c = 'k', linewidth = 0.5)
ax.plot()
plt.legend()
plt.savefig('./fig.png')
plt.show()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (17, 9))

ax.plot(cnn1[1500:], label = 'cnn1')
ax.plot(cnn2[1500:], label = 'cnn2')
plt.axhline(y = 0.0, c = 'k', linewidth=0.5)
ax.plot()
plt.legend()
plt.savefig('./fig.png')
plt.show()
```
